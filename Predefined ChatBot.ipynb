```python
import time
import re

def chatbot_response(user_input):
    # Converting the user input to lowercase for easier matching
    user_input = user_input.lower()
    
    # Rule-based responses
    if "hello" in user_input or "hi" in user_input:
        return "Hello! How can I assist you today?"
    
    elif "how are you" in user_input:
        return "I'm just a bot, but I'm doing great! How about you?"
    
    elif "your name" in user_input:
        return "I'm a simple rule-based chatbot created to help you."
    
    elif "bye" in user_input or "goodbye" in user_input:
        return "Goodbye! Have a great day!"
    
    elif "time" in user_input:
        current_time = time.strftime('%I:%M %p', time.localtime())
        return f"The current time is {current_time}."
    
    elif "date" in user_input:
        current_date = time.strftime('%B %d, %Y', time.localtime())
        return f"Today's date is {current_date}."
    
    elif "who created you" in user_input or "your creator" in user_input:
        return "I was created by a developer learning AI and rule-based systems!"
    
    elif "thank you" in user_input or "thanks" in user_input:
        return "You're welcome! Happy to help."
    
    elif "joke" in user_input:
        return "Why don't scientists trust atoms? Because they make up everything!"
    
    elif re.search(r'\badd\b|\bplus\b', user_input):  # Simple addition rule
        numbers = re.findall(r'\d+', user_input)
        if len(numbers) == 2:
            result = int(numbers[0]) + int(numbers[1])
            return f"The result is {result}."
        else:
            return "Please provide two numbers to add."
    
    elif re.search(r'\bsubtract\b|\bminus\b', user_input):  # Simple subtraction rule
        numbers = re.findall(r'\d+', user_input)
        if len(numbers) == 2:
            result = int(numbers[0]) - int(numbers[1])
            return f"The result is {result}."
        else:
            return "Please provide two numbers to subtract."
    
    elif re.search(r'\bdivide\b|\bdivision\b', user_input):  # Simple division rule
        numbers = re.findall(r'\d+', user_input)
        if len(numbers) == 2:
            if int(numbers[1]) == 0:
                return "Cannot divide by zero!"
            result = int(numbers[0]) / int(numbers[1])
            return f"The result is {result}."
        else:
            return "Please provide two numbers to divide."
    
    elif re.search(r'\bmultiply\b|\btimes\b', user_input):  # Simple multiplication rule
        numbers = re.findall(r'\d+', user_input)
        if len(numbers) == 2:
            result = int(numbers[0]) * int(numbers[1])
            return f"The result is {result}."
        else:
            return "Please provide two numbers to multiply."
    
    else:
        return "I'm sorry, I don't understand that. Could you please rephrase?"

# Main loop to keep the chatbot running
if __name__ == "__main__":
    print("Chatbot: Hello! Type 'bye' to exit.")
    while True:
        user_input = input("You: ")
        if user_input.lower() == "bye":
            print("Chatbot: Goodbye!")
            break
        response = chatbot_response(user_input)
        print(f"Chatbot: {response}")
```
